```python
import geopandas as gpd
import pandas as pd
from pathlib import Path
```

```python
# path
folder = Path('~/Documents/SIG/Otras capas venezuela/Areas_protegidas_WDPA_WDOECM_Mar2026_Public_VEN_shp').expanduser()

# merge 3 layers
abrae = gpd.GeoDataFrame(
    pd.concat([
        gpd.read_file(folder / 'WDPA_WDOECM_Mar2026_Public_VEN_shp_0'/'WDPA_WDOECM_Mar2026_Public_VEN_shp-polygons.shp'),
        gpd.read_file(folder /'WDPA_WDOECM_Mar2026_Public_VEN_shp_1'/ 'WDPA_WDOECM_Mar2026_Public_VEN_shp-polygons.shp'),
        gpd.read_file(folder /'WDPA_WDOECM_Mar2026_Public_VEN_shp_2'/ 'WDPA_WDOECM_Mar2026_Public_VEN_shp-polygons.shp'),
    ], ignore_index=True)
)
```

```python

# exploraration
pd.set_option('display.max_columns', None)
abrae.sample(5)
```

```python
# Extract only useful lyrs
abrae.columns
cols = ['SITE_ID','NAME_ENG','DESIG',
       'DESIG_ENG', 'IUCN_CAT','STATUS_YR', 'MANG_AUTH', 'MANG_PLAN', 'GIS_AREA','GIS_M_AREA','geometry']
```

```python
# New subset
abrae= abrae[cols]
abrae.sample(5)
```

```python
# Subset with only ABRAEs of interest
abrae['DESIG'].unique().tolist()
list_of_interest = ['Monumento Natural (MN)',
 'Parque Nacional (PN)', 'Reserva Nacional Hidraúlica (RNH)',
 'Zona Protectora (ZP)','Reserva Forestal (RF)', 'Área Boscosa Bajo Protección (ABBP)']
```

```python
abrae_clean = abrae[abrae['DESIG'].isin(list_of_interest)]
abrae_clean['DESIG'].value_counts()
```

```python
# Reproject ABRAE layers to the same Albers Equal-Area CRS used by the land cover rasters

# Output folder
output_path = Path.cwd() / 'abraes_lyrs'
output_path.mkdir(parents=True, exist_ok=True)

ALBERS = (
    "+proj=aea +lat_1=2 +lat_2=10 +lat_0=6 +lon_0=-66 "
    "+x_0=0 +y_0=0 +datum=WGS84 +units=m +no_defs"
)

# Ensure geometries are valid before reprojection
abrae_clean = abrae_clean.copy()
abrae_clean["geometry"] = abrae_clean.geometry.make_valid()

# Reproject to Albers
abrae_albers = abrae_clean.to_crs(ALBERS)

# Output
output_albers = output_path / "abraes_by_type_albers.gpkg"


# export each type in a layer within the geopackage
for type, group in abrae_albers.groupby('DESIG'):  
    lyr_name = type.replace(' ', '_').lower()
    group.to_file(output_albers, layer=lyr_name, driver='GPKG')

```
